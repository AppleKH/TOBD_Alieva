1.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

```python
content = open("steps_sample.xml","r").read()
soup = BeautifulSoup(content,'xml')


result_dict = {}
for recipe in soup.find_all('recipe'):
    recipe_id = int(recipe.find("id").get_text())
    result_dict[recipe_id] = []
    
    for step in recipe.find_all("step"):
        text = step.get_text()
        result_dict[recipe_id].append(text)

result_dict
```

1.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

```python
content = open("steps_sample.xml","r").read()
soup = BeautifulSoup(content,'xml')


result_dict = {}
for recipe in soup.find_all('recipe'):
    recipe_id = int(recipe.find("id").get_text())
    steps_count = len(recipe.find_all("step"))
    
    if steps_count not in result_dict:
        result_dict[steps_count] = []
    result_dict[steps_count].append(recipe_id)
        
result_dict
```

2.2 По заданному пользователем сдвигу задать систему кодирования букв (использовать код Цезаря). Сохранить систему кодирования в виде csv-файла.

```python
shift = int(input())  
a = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
b = []     
for c in a:
  c_index = ord(c) - ord("a")
  new_index = (c_index + shift) % 26
  new_unicode = new_index + ord("a")
  b.append(chr(new_unicode))
with open('code.csv', 'w', newline='')  as  f:
    fieldnames = ['Symbol', 'Encoded_symbol']
    writer = csv.DictWriter(f, fieldnames = fieldnames)
    writer.writeheader()
    for i in range(26):
      writer.writerow({'Symbol': a[i], 'Encoded_symbol': b[i]})
```

```python
with open('code.csv', newline='') as f:
    data = csv.reader(f, delimiter=';')
    for row in data:
        print(row)
```

2.3	На основе сохраненной системы кодирования кодировать и декодировать введенное пользователем сообщение.

```python
message = input()
en_message = ''
d_message = ''
chars = []
with open('code.csv', newline='') as f:
    data = csv.reader(f, delimiter=';')
    for row in data:
      chars.append(row)
for c in message:
  for row in chars:
    row = row[0].split(',')
    if row[0] == c:
      en_message += row[1]
for c in en_message:
  for row in chars:
    row = row[0].split(',')
    if row[1] == c:
      d_message += row[0]
print(en_message)
print(d_message)
```
