1. Постройте график линейной функции (kx+b) с коэффициентами k и b, вводимыми пользователем. Предварительно рассчитать 100 значений аргумента из диапазона введенного пользователем.


```python
d1, d2 = map(int, input().split())
k, b = map(int, input().split())
x = []
y = []
for i in np.linspace(d1,d2,100):
  x.append(i)
  y.append(k*i+b)
plt.plot(x,y)
```

2. Постройте графики y = sin(x) и y = e-0,1xcos(x) на одной картинке. Сделайте их разными цветами и добавьте легенды.

```python
x = np.linspace(-10,10,100)
y1 = np.sin(x)
y2 = np.exp(-0.1 * x) * np.cos(x)
plt.plot(x,y1,label='y=sin(x)')
plt.plot(x,y2,label='y=exp(-0.1x)cos(x)')
plt.legend()
```

3. Постройте графики y = x2 и y = |x| на двух картинках в строке. Добавьте заголовки и подписи осей.

```python
y1 = [i*i for i in x]
y2 = [abs(i) for i in x]
fig, ax = plt.subplots(1,2, figsize=(20,11))
ax[0].plot(x,y1)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y1')
ax[0].set_title('y1 = x^2')
ax[1].plot(x,y2)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y2')
ax[1].set_title('y2 = |x|')
plt.show()
```

6. Сгенерируйте случайный процесс X(t) = r*X(t-1) + e(t), где дисперсия случайной нормальной добавки De = (1-r^2)*Dx. Dx = 1, r = 0.93. Отобразите его на графике.


```python
N = 1000
Dx = 1
r = 0.9
De = (1 - r**2) * Dx
x = np.zeros(N)
x[0] = np.random.normal(0, np.sqrt(Dx))
for i in range(1,N):
  x[i] = x[i-1] * r + np.random.normal(0,np.sqrt(De))
df = pd.DataFrame({'x':x})
plt.plot(df)
plt.title('x(t) = x(t-1)*r + e(t)')
plt.grid(color='black', alpha = 0.8, linewidth=1)
plt.show()
```

9. Построить секторную диаграмму для X из задания 6. Секторами диаграммы должно быть количество значений в интервале: [-inf, -2], [-2,0], [0, inf]. 


```python
values = df['x']
fig, ax = plt.subplots()
interval = [0, 0, 0]

labels = ['[-inf, -2]',
          '[-2, 0]',
          '[0, +inf]']

for i in range(0, len(values)):
  if (values[i] <= -2):
    interval[0]+= 1
  elif (values[i] >= 0):
    interval[2] += 1
  else:
    interval[1] += 1

ax.pie(interval, autopct = '%1.2f%%', labels = labels, colors = ['#e2d810', '#d72631','#5c3c92'])
ax.axis("equal")
```
