```python
#1 Создайте две серии данных Pandas. В первой будут храниться значения в диапазоне [1; 100], а во второй случайные целочисленные значения (равномерное распределение) из диапазона [0,10]. Объедините их в DataFrame.
import numpy as np
import pandas as pd

ser1 = pd.Series(np.arange(1,101))
ser2 = pd.Series(np.random.randint(0,11,size=100))
df = pd.concat([ser1,ser2],axis=1)
df.rename(columns={0:'s1', 1:'s2'}, inplace=True)
df.head()
```

```python
#2 Выведите описательную статистику по имеющимся сериям, где в качестве строк выступают названия столбцов.
df.describe().T
```

```python
#3 Из обеих колонок отнимите среднее значение и поделите на стандартное отклонение (нормализуйте данные). Повторите вывод описательной статистики. 
df = (df-df.mean(axis=0))/(df.std(axis=0))
df.describe()
```

```python
#4 Сделайте масштабирование в каждом столбце в диапазоне от 0 до 1 и выведите описательную статистику, не меняя исходных данных.
df1 = df.copy()
df1 -= df1.min(axis=0)
df1 /= df1.max(axis=0)
df1.describe()
```

```python
#5 Добавьте столбец s3, вычисляемый как произведение столбцов s1 и s2. 
df.insert(2,'s3',df['s1']*df['s2'])
df
```

```python
#6 Сохраните данные в csv-файл, а затем считайте их из него.
df.to_csv('file.csv', index=False)
dfd = pd.read_csv('file.csv')
dfd.head(3)
```

```python
#7 Выберите из прочитанного датасета только первый и третий столбцы, у которых значение в третьем столбце положительно.
df.iloc[:,[0,2]][df.iloc[:,2]>0]
```

```python
#8 Отсортируйте данные по убыванию значений в третьем столбце.
df.sort_values(by=['s3'],ascending=False)
```

```python
#9 Добавьте столбец 4, как линейную сумму столбца 3 с некоторыми заданными коэффициентами (ax+b) и случайной добавкой (r). Определите корреляцию между всеми столбцами. r - равномерное число [0,1).
a = 1
b = 2
dfd.insert(3,'s4',dfd['s3']*a+np.random.rand())
dfd
```

```python
#10 Найдите евклидово расстояние между двумя произвольными Series (точками) a и b? При этом a и b соответствуют случайно выбранным строкам из DataFrame.
sam = df.sample(2)
np.linalg.norm(np.array(sam.iloc[0]) - np.array(sam.iloc[1]))
```
